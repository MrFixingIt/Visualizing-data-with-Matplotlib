## Visualizing Data with Pandas and MatPlotLib


### As a Data Analyst, I want to set up the proper imports so I have access to the Pandas and MatPlotLib libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### As a Data Analyst, I want to import and store the `ufo_sightings.csv` file in a variable

```python
ufo_sightings_df = pd.read_csv("ufo_sightings.csv")
print(ufo_sightings_df)
```

### As a Data Analyst, I want to ensure my DataFrame is cleaned by removing any **null** values

```python
nulls_removed_df = ufo_sightings_df.dropna()
nulls_removed_df
```

### As a Data Analyst, I want to filter my DataFrame to remove any information that is not related to my analysis
* This will include **dropping** the **Described_Duration, Description, and Date_Documented** columns

```python
columns_removed_df = nulls_removed_df.drop(['Described_Duration', 'Description', 'Date_Documented'], axis=1)
columns_removed_df
```

### As a Data Analyst, I want to convert the `Date_Time` column to utilize Pandas' DateTime data type

```python
columns_removed_df["Date_Time"]= columns_removed_df["Date_Time"].astype('datetime64[ns]')
columns_removed_df
```

## Data Visualization with Matplotlib

### As a Data Analyst, I want to visualize the **frequency of different UFO shapes**, to determine which UFO shapes are most commonly encountered
* I will use a **Bar Graph** to visualize this data
* I will include a label for the X and Y axis
* I will include a name for my chart

```python
# title = "Frequency of Different UFO Shapes"
# print(title)
bar_graph_df = columns_removed_df["UFO_Shape"].value_counts()
bar_graph_df.plot.bar()
plt.title('Frequency of Different UFO Shapes')
```

### As a Data Analyst, I want to visualize the **number of UFO encounters**, grouped by each decade, to see if there is a correlation between number of sightings and decade
* I will use a **Bar Graph** to visualize this data
* I will include a label for the X and Y axis
* I will include a name for my chart

```python
columns_removed_df["Date_Time"] = pd.to_datetime(columns_removed_df["Date_Time"])
columns_removed_df["Year"] = columns_removed_df["Date_Time"].dt.year
columns_removed_df["10 Year Interval"] = (columns_removed_df["Year"] // 10) * 10
# columns_removed_df["Month"].value_counts().sort_index()
bar_graph_encounters_df = columns_removed_df["Year"].value_counts().sort_index()
bar_graph_encounters_df.plot.bar(figsize=(12,4))
plt.title('Number of UFO Encounters by Decade')

# columns_removed_df.info()
# columns_removed_df.head()
# columns_removed_df.plot.bar()
# 11:32 timestamp Plotting Dates
```

### As a Data Analyst, I want to create a **Line Graph** to visualize the **Frequency of UFO sightings, by hour of the day**
* I will use a **Line Graph** to visualize this data
* I will include a label for the X and Y axis
* I will include a name for my chart

```python
columns_removed_df["Date_Time"] = pd.to_datetime(columns_removed_df["Date_Time"])
columns_removed_df["Hour of the Day"] = columns_removed_df["Date_Time"].dt.hour
bar_graph_encounters_df = columns_removed_df["Month"].value_counts().sort_index()
bar_graph_encounters_df.plot.line()
plt.title('Frequency of UFO Sightings by Hour of Day')
```
